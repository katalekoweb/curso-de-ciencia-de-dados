```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("trees.csv")
base.shape
```

```python
# Geracao de boxplot
# patch_artist = True preenche showfliers outliers
plt.boxplot(base.Volume, vert=False, showfliers=False, notch=True,patch_artist=True)
plt.title("Arvores")
plt.xlabel("Volume")
```

```python
# Dados por linha
plt.boxplot(base)
plt.title("Arvores")
plt.xlabel("Dados")
```

```python
# geracao de 3 boxplots, cada um mostrando informacoes diferentes
plt.boxplot(base.Volume, vert=False)
plt.boxplot(base.Girth, vert=False)
plt.boxplot(base.Height, vert=False)
plt.title("Arvores")
plt.xlabel("Dados")
```
