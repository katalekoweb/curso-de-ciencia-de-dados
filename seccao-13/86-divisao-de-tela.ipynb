```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Carregamento dos dados
base = pd.read_csv("trees.csv")
base.head()
```

```python
base.shape
```

```python
# girth com volume
plt.scatter(base.Girth, base.Volume)
```

```python
# girth com height
plt.scatter(base.Girth, base.Height)
 
```

```python
# height com volume
plt.scatter(base.Height, base.Volume)
```

```python
# histograma do volume
plt.hist(base.Volume)
```

```python
# Imprimindo juntos
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker = "*")
plt.subplot(2,2,4)
plt.hist(base.Volume)
```
